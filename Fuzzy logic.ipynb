Данные получены из статьи статьи Бородкин Л.И. Дискриминантный анализ с использованием концепции нечетких множеств в задаче оценки двух путей аграрной эволюции России на рубеже XIX–XX вв. // Историческая информатика. 2013. № 3. С. 30-38

```python
import pandas as pd
import numpy as np
from fcmeans import FCM

df = pd.read_excel('test.xlsx')
df
```

```python
X = df.drop('Region', axis = 1)
X = X.to_numpy()
```

```python
# запускаем модель
fcm = FCM(n_clusters=2, tolerance = 0.5) 
fcm.fit(X) 
```

```python
centers = fcm.centers
labels = fcm.predict(X)
```

```python
np.set_printoptions(suppress=True)
display(centers)
```

```python
# вес принадлежности к типу
res = pd.DataFrame()
res['Region'] = df['Region']
res[['Type_1','Type_2']] = (fcm.u)
res['Type_1'] = res['Type_1'].round(decimals = 2)
res['Type_2'] = res['Type_2'].round(decimals = 2)
res['Result u = 0.50'] = labels
```

```python
res_init = pd.read_excel('Book 1.xlsx')
res['Init Type_1'] = res_init['Peasant type']
res['Init Type_2'] = res_init['Proprietor type']
res['Result u = 0.70'] = res['Type_1'].apply([lambda x: 0 if x > 0.7 else 1])
```

```python
col_list = res.columns
col_list
new_order = ['Region', 'Type_1', 'Init Type_1','Type_2', 'Init Type_2', 'Result u = 0.50', 'Result u = 0.70']
res = res[new_order]

res
```
